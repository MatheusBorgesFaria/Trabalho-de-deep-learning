notebook usado para filtrar modelos de interesse e selecionar colunas SMART

```python
import os
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
csv_files = [arquivo for arquivo in os.listdir('../data/data_q3_2024') if arquivo.startswith("2024-07") & arquivo.endswith(".csv")]
csv_files
```

```python
n_columns = [1, 5, 7, 184, 187, 188, 189, 190, 193, 194, 197, 198, 240, 241, 242]
normalized_cols = [f'smart_{n}_normalized' for n in n_columns]
#raw_cols = [f'smart_{n}_raw' for n in n_columns]
```

```python
df = pd.DataFrame()
for file in csv_files:
    print(file)
    df_aux = pd.read_csv(f'../data/data_q3_2024/{file}')
    id_cols = [c for c in df_aux.columns[:11]]
    cols = id_cols + normalized_cols
    df_aux = df_aux[df_aux['model'].isin(['ST16000NM001G', 'ST12000NM0008'])][cols]
    df = pd.concat([df, df_aux])

print(df.shape)
df.head()
```

```python
df.to_csv('../data/processed/jul_2024.csv', index=False)
```
