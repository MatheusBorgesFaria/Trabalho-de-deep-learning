```python
import os
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

**Dataset completo overview**   
 Quais variáreis estão preenchidas e possuem valores distintos? Quantos modelos? Tem um modelo com maior taxa de erro? Quantos discos? Todos os discos tem a mesma temporalidade de registros? Quando tem erro, o erro permanece por quanto tempo? 

```python
# carrega dados de setembro
path_raw_data = '../data/data_q3_2024'
csv_files = [arquivo for arquivo in os.listdir(path_raw_data) if arquivo.startswith("2024-09-")]
df = pd.DataFrame()
for file in csv_files:
    print(file)
    df_aux = pd.read_csv(f'{path_raw_data}/{file}')
    df = pd.concat([df, df_aux])

print(df.shape)
df.head()
```

___
Características gerais

```python
# lista variáveis no csv original e principais informações

def summarize_dataframe(df):
    summary = pd.DataFrame({
        'Column': df.columns,
        'Distinct Values': [df[col].nunique() for col in df.columns],
        'Null Values': [df[col].isnull().sum() for col in df.columns],
        '% Values': [df[col].notnull().sum()/len(df) for col in df.columns],
        'Data Type': [df[col].dtype for col in df.columns]
    })
    return summary

# Resumo do DataFrame
summary = summarize_dataframe(df)
print(summary)
```

Algumas variáveis tem muitos valores nulos ou um valor constante. 
Exemplo 198 possui menos que 10% dos valores preenchidos.

```python
# resumo informções por modelo

df_model = df.groupby('model').agg({'date':['min','max'], 'serial_number':['nunique', 'count'], 'failure':'sum'}).reset_index()
df_model.columns = ['model', 'first_date', 'last_date', 'disks', 'records', 'failure_records']
df_model['failure_rate'] = df_model['failure_records']/df_model['records']*100
df_model.sort_values('failure_rate', ascending=False)
```

Há uma variação no número de discos e registros por modelo. A taxa de falha também muda de acordo com o modelo.

```python
# Há discos que não aparecem nos 30 dias do mês?
df.serial_number.value_counts().reset_index().sort_values('count').head(10)
```

```python
# Quantos discos que não aparecem nos 30 dias do mês?
len(df.serial_number.value_counts().reset_index()[df.serial_number.value_counts().reset_index()['count']<30])
```

```python
# Exemplo: periodicidade e indicativa de falha para discos que aparecem 25 dias 

teste_25dias = df.serial_number.value_counts().reset_index()[df.serial_number.value_counts().reset_index()['count']==25].head(10).serial_number
df[df.serial_number.isin(teste_25dias)].groupby('serial_number').agg({'date':['min','max'], 'failure':['count','sum']}).reset_index()
```

As vezes os registros começam depois, as vezes param antes (nos exemplos acima, os que param antes do final do mês, falharam)

___
Explorando erros

```python
# Total de falhas  
print('total de falhas:', df['failure'].sum())
```

```python
# Quantos discos falharam mais de uma vez?
len(df.groupby('serial_number')['failure'].sum().reset_index()[df.groupby('serial_number')['failure'].sum().reset_index()['failure']>1])
```

```python
# o último registro é o dia da falha?

# Determina a data máxima por ID
max_date_disk = df.groupby(['model','serial_number'])['date'].max().reset_index()
max_date_disk.rename(columns={'date': 'max_date'}, inplace=True)

# Determina a data de falha por ID (caso exista falha=1)
failure_date = df[df['failure'] == 1].groupby('serial_number')['date'].max().reset_index()
failure_date.rename(columns={'date': 'failure_date'}, inplace=True)

# Junta os dois resultados no dataframe original
output = pd.merge(max_date_disk, failure_date, on='serial_number', how='right')

# Adiciona uma coluna indicando se há registros após a falha
print(len(failure_date))
print(len(output[output.max_date > output.failure_date]))
output[output.max_date > output.failure_date]
```

351 falhas ao todo. Nenhum disco falha mais de uma vez. Sete casos que tem registro depois da falha (não necessariamente continuam operando até o final do mês). 

```python
# exibindo casos que falham e continuam operando

df[df.serial_number.isin(output[output.max_date > output.failure_date].serial_number)].sort_values(['serial_number','date'])
```
